```python
xrun_list = []
yrun_list = []
trun_list = []

for x,values in c1.items():
    for y,obj in values.items():
        obj.run(timeout = 90)
        t = obj.get_runtime()
        
        print(f'x = {x}, y = {y}, t = {t}')
        xrun_list.append(x)
        yrun_list.append(y)
        trun_list.append(t)
        
        if t == -1:
            break
```

```python
x_list = []
y_list = []
t_list = []

for x,values in c1.items():
    for y,obj in values.items():
        t = obj.tp_time
        print(f'x = {x}, y = {y}, t = {t}')
        x_list.append(x)
        y_list.append(y)
        t_list.append(t)
```
